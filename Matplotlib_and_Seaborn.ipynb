```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
print(tips.head())


tips_by_day = tips.groupby("day")["tip"].sum()

plt.figure(figsize=(8, 5))
plt.bar(tips_by_day.index,
        tips_by_day.values)


plt.title("Total Tips by Day")
plt.xlabel("Day")
plt.ylabel("Total Tips")
plt.show
```

```python
#-------------Task 2: ----------------


plt.figure(figsize=(8, 5))

sns.histplot(data=tips, x="tip", bins=20, kde=True)

plt.title("Distribution of Tip amount")
plt.xlabel("Tip Amount")
plt.ylabel("count")
plt.tight_layout()
plt.show()

# most tip amounts b/w 2$ and 4$
# KDE line show  the histogram trand
```

```python
#------------Task 3: ------------------

plt.figure(figsize=(8, 5))

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", alpha=0.8)

plt.title("Total Bill vs Tip Amount")
plt.xlabel("Total Bill")
plt.ylabel("tip")
plt.tight_layout()
plt.show()

# Higher bill lead to higher tips and in some cases variation exists
```
