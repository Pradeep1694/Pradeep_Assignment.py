```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")

#----------Task 1:Basic Grouping and Single Aggregations------------

sales_by_region = df.groupby('region')['sales_amount'].sum().reset_index()
print("\nTotal Sales by Region")
print(sales_by_region)


Transactions_by_category = df.groupby('category')['transaction_id'].count().reset_index()
print("\nTransactions by Category")
print(Transactions_by_category)


avg_sales_salesperson = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nAverage Sales by Salesperson")
print(avg_sales_salesperson)


sales_by_region = df.groupby('region')['sales_amount'].sum().sort_values(ascending=False)
print(f"\nRegion sales in descending order")
print(sales_by_region)


#----------Task 2: Multi-Column Grouping ----------

region_category = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
print(f"\nTotal Sales by Region and Category")
print(region_category)

salesperson = df.groupby('salesperson').agg({'sales_amount': ['sum', 'mean', 'count']}).reset_index()
print("\nSalesperson Summary")
print(salesperson)


top_salesperson = df.groupby('salesperson')['sales_amount'].sum().sort_values(ascending=False)
print("\nTop Performing Salesperson")
print(top_salesperson)


best_category = df.groupby('category')['sales_amount'].sum().idxmax()
print("\nCategory with Maximum Revenue:",best_category)



#----------Task 3: Custom Aggregation and Complete Sales Report-------------

def sales_range(x):
    return x.max() - x.min()

result = df.groupby('region')['sales_amount'].agg(sales_range).reset_index()
print("\nSales Range by Region")
print(result)

region_sales = df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print(f"\nRegion and sales range")
print(region_sales)


final_summar = df.groupby('region').agg({'sales_amount': ['sum', 'mean'],'customer_id': 'count'}).reset_index()
print(f"\nFinal Summary")
print(final_summar)


```
