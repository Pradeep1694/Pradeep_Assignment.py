```python
import requests

import pandas as pd

# API URL
url = "https://restcountries.com/v3.1/all"

# -------------------------------
# Task 1: Fetch and Validate Response
# -------------------------------
try:
    response = requests.get(url)
    print("HTTP Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries:", len(data))
    else:
        print("Error: Failed to fetch data")

except Exception as e:
    print("Exception occurred:", str(e))

#-----------Task 2 -----------------

countries_list = []

for country in data:
  try:
      name = country.get("name, {}").get("common", "N/A")
      population = country.get("population", 0)
      region = country.get("region", "N/A")
      area = country.get("area", 0)

      countries_list.append({
          "Name": name,
          "Population": population,
          "Region": region,
          "Area": area
      })
  except Exception:
        continue

df = pd.DataFrame(countries_list)
print("\nFirst 10 rows)")
print(df .head(10))


#--------------Task 3 -------------

region_population = df.groupby("region")["population"].sum()
highest_population_region = region_population.idxmax()

largest_country = df.loc[df["area"].idxmax(), "name"]

print("\nSummary Results:")
print("Region with highest population:", highest_population_region)
print("Country with largest area:", largest_country)
```
