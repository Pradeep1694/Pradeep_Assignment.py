```python
import pandas as pd


students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

#-----------Task 1:
# Create DataFrames

students_df = pd.DataFrame(students_data)
enrollments_df = pd.DataFrame(enrollments_data)
scores_df = pd.DataFrame(scores_data)

print("Students DataFrame")
print(students_df)

# Display null count and percentage
print("\nMissing values count")
print(students_df.isnull().sum())
print("\nMissing values percentage")
print(students_df.isnull().mean() * 100)

# Fill missing city
students_df["city"] = students_df["city"].astype("category")
students_df["city"] = students_df["city"].cat.add_categories("Unkown")
students_df["city"] = students_df["city"].fillna("Unkown")

# Drop rows where name is missing
students_df = students_df.dropna(subset=["name"])

print("\ncleaned students DataFrame")
print(students_df)

#----------Task 2: Join Operations----------

inner = pd.merge(students_df, enrollments_df, on="student_id", how="inner")
print("\nInner Join")
print(inner)
print("Students in result:", len(inner))

left = pd.merge(students_df, enrollments_df, on="student_id", how="left")
print("\nLeft Join")
print(left)
print("Total rows:", len(left))

right = pd.merge(students_df, enrollments_df, on="student_id", how="right")
print("\nRight Join")
print(right)
print("Total rows:", len(right))

outer = pd.merge(students_df, enrollments_df, on="student_id", how="outer")
print("\nOuter Join")
print(outer)
print("Total rows:"),len(outer)


# Show rows with null values
print("\nRows with missing name or course:")
print(outer[(outer['name'].isnull()) | (outer['course_name'].isnull())])

# Indicator merge
indicator = pd.merge(students_df, enrollments_df, on='student_id', how='outer', indicator=True)
print("\nMerge Source Distribution:")
print(indicator["_merge"].value_counts())

#---------- Task 3: Lookup Operation----------
#Create dictionary for mapping

score_lookup = dict(zip(scores_df["student_id"], scores_df["exam_score"]))
print("\n lookup dict:")
print(score_lookup)

# Add exam score using map
students_df['exam_score'] = students_df['student_id'].map(score_lookup)
print("\nStudents with Scores (using map):")
print(students_df)

#Performance Comparison

merge_scores = pd.merge(students_df, scores_df, on="student_id", how="left")
print("\nScores using merge:")
print(merge_scores)

#Simple Automation

def auto_merge(df1, df2, join_type, key_column):
    merged_df = pd.merge(df1, df2, on=key_column, how=join_type)
    result = {
        "merged_df": merged_df,
        "row_count": len(merged_df),
        "join_type": join_type
    }
    return result

# Testing function
test1 = auto_merge(students_df, enrollments_df, "inner", "student_id")
test2 = auto_merge(students_df, enrollments_df, "left", "student_id")

print("\nAuto Merge Test 1:")
print(test1)

print("\nAuto Merge Test 2:")
print(test2)



```
