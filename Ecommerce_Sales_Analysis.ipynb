```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")

#----------Task 1:------------

sales_by_region = df.groupby('region')['sales_amount'].sum().reset_index()

# transactions for each product category

Transactions_by_category = df.groupby('category')['transaction_id'].count().reset_index()

avg_sales_salesperson = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print("\nTotal Sales by Region")
print(sales_by_region)

print("\nTransactions by Category")
print(Transactions_by_category)

print("\nAverage Sales by Salesperson")
print(avg_sales_salesperson)
```
