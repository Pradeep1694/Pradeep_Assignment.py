```python
import pandas as pd
import sqlite3

customers_url = "https://raw.githubusercontent.com/graphql-compose/graphql-compose-examples/master/examples/northwind/data/csv/customers.csv"
orders_url = "https://raw.githubusercontent.com/graphql-compose/graphql-compose-examples/master/examples/northwind/data/csv/orders.csv"

customers_df = pd.read_csv(customers_url)
orders_df = pd.read_csv(orders_url)

conn = sqlite3.connect(":memory:")
customers_df.to_sql("customers", conn, index=False, if_exists="replace")
orders_df.to_sql("orders", conn, index=False, if_exists="replace")


#----------Task 1: ----------------


query="""
SELECT
    CustomerID,
    COUNT(OrderID) AS order_count,
    SUM(Freight) AS total_freight,
    AVG(Freight) AS avg_freight
FROM orders
GROUP BY CustomerID
ORDER BY total_freight DESC
LIMIT 10;
"""

query = pd.read_sql_query(query, conn)
display(query)
```

```python
#-----------Task 2 A: --------------

query_a ="""
SELECT CustomerID, COUNT(OrderID) AS high_freight_orders
FROM orders
WHERE Freight > 50
GROUP BY CustomerID
"""
query_a = pd.read_sql_query(query_a, conn)
display(query_a)
```

Query filters individual rows based on the Freight value before calculation happens

```python
#-----------Task 2 B:------------

query_b ="""
SELECT CustomerID, SUM(Freight) AS total_freight
FROM orders
GROUP BY CustomerID
HAVING total_freight > 500;
"""
query_b = pd.read_sql_query(query_b, conn)
display(query_b)
```

Query groups all orders by customer first calculates the total and then filters out customers total spending is below 500

```python
#-----------Task 3:-------------
inner_join_query = """
SELECT
c.CompanyName,
c.Country,
COUNT(o.OrderID) AS order_count,
SUM(o.Freight) AS total_freight
FROM customers c
INNER JOIN orders o ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID;
"""
inner_join_query = pd.read_sql_query(inner_join_query, conn)
display(inner_join_query)
```

Inner join only returns customers who have placed at least one order

```python
#------------Task 3: -------------

left_join_query = """
SELECT
c.CompanyName,
c.Country,
COUNT(o.OrderID) AS order_count,
SUM(o.Freight) AS total_freight
FROM customers c
LEFT JOIN orders o ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID;
"""
left_join_query = pd.read_sql_query(left_join_query, conn)
display(left_join_query)
```

Left join returns every customer in the database those who have not placed an order and order count will be 0 and total freight will be NaN
